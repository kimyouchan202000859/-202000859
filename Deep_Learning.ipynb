```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
targetUrl_iris="https://raw.githubusercontent.com/kimyouchan202000859/python/main/caffeine.csv"
df = pd.read_csv(targetUrl_iris)
x = df.iloc[:,1:4]
std = StandardScaler()
x = std.fit_transform(x)
y = df.iloc[:,4]
y = pd.get_dummies(y)
print(y[0:5])

model = Sequential()
model.add(Dense(128, input_dim=3, activation='relu'))  #input, dense, activation 설정
model.add(Dense(82, activation='relu'))
model.add(Dense(70, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(6, activation='softmax'))
model.summary()

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history=model.fit(x, y, epochs=30000, batch_size=256)
y_accu = history.history['accuracy']
x_len = np.arange(len(y_accu))
plt.plot(x_len,y_accu)
plt.show()
```
